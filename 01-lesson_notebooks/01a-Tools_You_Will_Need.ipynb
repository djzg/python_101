## Setup your local git global settings

Thanks to [Earth Lab](https://www.earthdatascience.org/courses/intro-to-earth-data-science/git-github/version-control/fork-clone-github-repositories/) for their comprehensive version control setup instructions. A summary of setup from them is below.

The first time that you use git on a computer, you will need to configure your GitHub.com username and email address. This information will be used to document who made changes to files in git. It is important to use the same email address and username that you setup on GitHub.com.

You can set your Github.com username in the terminal by typing:

`$ git config --global user.name "username".`

Next, you can set the email for your Github.com account by typing:

`$ git config --global user.email "email@email.com".`

Using the `--global` configuration option, you are telling git to use these settings for all git repositories that you work with on your computer. Note that you only have to configure these settings one time on your computer.

You can check your config settings for `user.name` and `user.email` using the following commands:

`git config user.name` which returns the username that you set previously

`git config user.email` which returns the email that you set previously

These configuration settings ensure that changes you make to repositories are attributed to your username and email.

Find those instructions [here](https://www.earthdatascience.org/courses/intro-to-earth-data-science/git-github/version-control/how-to-setup-git/) (you will need to scroll down the page).

## Fork the class materials
From this project's main repository: https://github.com/branhoff/python_class_101.git, select the "fork" option to create a copy of this project under your GitHub profile. The directory for this forked repository will be the equivalent of `https://github.com/[your.user.name]/python_class_101.git`

## Create projects directory
Create  directory on your machine where you want to house future projects you want to save in GitHub (generally programming projects).

In your terminal you can type `cd ~` to go to your "home directory". I advise you create a directory called "GitProjects" or "Projcets".

You can do that from the command line by typing `mkdir GitProjects`. If you type `ls` on the command line you should see the directory you created listed.

You can also type `pwd` (present working directory) to see where you've created your project if you're not sure.

Or naturally, create all this manually through your computer's standard interface.

## Clone the class materials
The latest materials are saved at github at this location: https://github.com/branhoff/python_class_101

Let's clone from the forked directory you made. We can easily access them by using the `git clone https://github.com/[your.user.name]/python_class_101.git` command.

If you followed the steps for setting up `ssh` keys for github (which I do recommend), you'll use a command like `git clone git@github.com:[your.user.name]/python_class_101.git` instead.

## Cloning for Windows
In the directory you created to house git related projects (I like the name GitProjects), right click on your created directory, and select your `Git bash here` option.

A git bash window will open up at the file location. Now simply type in the command `git clone https://github.com/[your.user.name]]/python_class_101.git` or `git clone git@github.com:[your.user.name]/python_class_101.git` if you used ssh.

You should now see in your created directory a new directory called `python_class_101` with all of the material for the course.

## Install Jupyter Notebook for Windows
We can use python's `pip` (Python Package Index) to install libraries and dependencies that don't come included with Python's standard library. In our case, we need the open-source tool Jupyter Notebook (introduced in *01b-Some_Context*).

Open up your Git Bash window and type the command `pip install jupyter`

## Install Jupyter Notebook for Mac
